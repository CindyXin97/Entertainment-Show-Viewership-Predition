```python
from __future__ import print_function
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
%matplotlib

def add_layer(inputs, in_size, out_size, activation_function=None):
    Weights = tf.Variable(tf.random_normal([in_size, out_size]))
    biases = tf.Variable(tf.zeros([1, out_size]) + 0.1)
    Wx_plus_b = tf.matmul(inputs, Weights) + biases
    Wx_plus_b = tf.nn.dropout(Wx_plus_b, keep_prob)
    if activation_function is None:
        outputs = Wx_plus_b
    else:
        outputs = activation_function(Wx_plus_b)
    return outputs,Weights

def add_layer_out(inputs, in_size, out_size, activation_function=None):
    Weights = tf.Variable(tf.random_normal([in_size, out_size]))
    biases = tf.Variable(tf.zeros([1, out_size]) + 0.1)
    Wx_plus_b = tf.matmul(inputs, Weights) + biases
    if activation_function is None:
        outputs = Wx_plus_b
    else:
        outputs = activation_function(Wx_plus_b)
    return outputs,Weights

# Make up some real data
# noise = np.random.normal(0, 1, np.arange(0,1000,1).shape)
# x_data=np.ones((1000,5))
# x_data[:,0]=np.arange(0,1000,1)+100
# x_data[:,1]=np.arange(0,1000,1)/2
# x_data[:,2]=np.arange(0,1000,1)/5
# x_data[:,3]=np.arange(0,1000,1)*2
# x_data[:,4]=np.arange(0,1000,1)*2
# y_data=np.zeros((1000,1))
# y_data[:,0]=np.square(np.arange(0,1000,1)) + noise * 40
# for j in range(1000):
#     if j%2==1:
#         y_data[j,0]=0
#     else:
#         y_data[j,0]=1

# define placeholder for inputs to network
keep_prob = tf.placeholder(tf.float32)
xs = tf.placeholder(tf.float32, [None, 5])
ys = tf.placeholder(tf.float32, [None, 1])
# add hidden layer
l1,w1 = add_layer_out(xs, 5, 50, activation_function=tf.nn.relu)
l2,w2= add_layer_out(l1, 50, 50, activation_function=tf.nn.relu)
# add output layer
prediction,w_out = add_layer_out(l2, 50, 1, activation_function=None)

#Loss function
#loss = tf.reduce_mean(tf.reduce_sum(tf.square(ys-prediction), reduction_indices=[1]))
#loss = tf.reduce_mean((tf.square(ys-prediction)))

#loss=tf.reduce_mean(tf.squared_difference(ys, prediction))
#loss=tf.reduce_mean(tf.squared_difference(ys, prediction))
loss=tf.reduce_mean(tf.squared_difference(ys, prediction))+0.5*(tf.reduce_sum(tf.square(w1))+tf.reduce_sum(tf.square(w2)))
loss_2=tf.square(ys-prediction)

#Optimizer
#train_step = tf.train.GradientDescentOptimizer(0.1).minimize(loss)
#train_step=tf.train.MomentumOptimizer(0.05, 0.05).minimize(loss)
train_step=tf.train.RMSPropOptimizer(0.01).minimize(loss)



#Training
sess = tf.Session()
sess.run(tf.global_variables_initializer())

##plot the real data
# fig = plt.figure()
# ax = fig.add_subplot(1,1,1)
# ax.scatter(np.arange(0,1000,1), y_data,s=3)
# plt.ion()
# plt.show()
ans=[]
p=[]
for i in range(10000):
    # training
    #print((sess.run(prediction, feed_dict={xs: x_data, ys: y_data})))
    sess.run(train_step, feed_dict={xs: x_data, ys: y_data,keep_prob: 1})
    #print(sess.run(prediction, feed_dict={xs: x_data, ys: y_data}))
    #print('jjjjjjjjjjj')
    
    if i % 50 == 0:
        # to visualize the result and improvement
#         try:
#             ax.lines.remove(lines[0])
#         except Exception:
#             pass
        prediction_value = sess.run(prediction, feed_dict={xs: x_data, ys: y_data,keep_prob:1})
        # plot the prediction
        p.append(prediction_value)
#         lines = ax.scatter(np.arange(0,10,1), prediction_value)
#         plt.pause(1)
        ans.append(sess.run(loss, feed_dict={xs: x_data, ys: y_data,keep_prob: 1}))

#plt.scatter(list(range(1,len(ans)+1)),ans)
```
